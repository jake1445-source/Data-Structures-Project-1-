## B. Sorting and Performance Testing 

Below is the code to populate the events instance with $n$ events. These events are then sorted using the 3 different algorithms. Then some plotting code will display the time taken to sort the lists using each algorithm across a number of trials with an increasing number of events. 

```python
events = generate_events(n = 50)


# # Uncomment to ensure sorting works 

# insertion_sort(events.data)
# merge_sort(events.data)
# quick_sort(events.data)
```

```python


trials = [10,50,100, 500,1000, 3000, 5000, 10000]
runtimes_insertion = []
runtimes_merge = []
runtimes_quick = []

for n in trials:
  s1 = runtime(n,insertion_sort)
  runtimes_insertion.append(s1)
  s2 = runtime(n,merge_sort)
  runtimes_merge.append(s2)
  s3 = runtime(n,quick_sort)
  runtimes_quick.append(s3)

plt.figure(figsize=(6,6))
plt.semilogy(trials,runtimes_insertion, marker = 'o', label = 'Insertion Sort')
plt.semilogy(trials,runtimes_merge, marker = 'x', label = 'Merge Sort')
plt.semilogy(trials,runtimes_quick,marker='*',label ='Quick Sort')
plt.legend()
```
